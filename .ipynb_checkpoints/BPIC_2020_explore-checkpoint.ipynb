## Проверяем наличие одинаковых аттрибутов 

```python
data = [dom_decl_df, int_decl_df, perm_log_df, req_paym_df, prep_trv_df]
all_col = []
imgs = []
shared_cols = []
for i in range(len(data)-1):
    for j in range(len(data)-i-1):
        all_col.append(list(data[i].columns))
        all_col[len(all_col)-1] += list(data[i+j+1].columns)
        all_col[len(all_col)-1] = np.unique(all_col[len(all_col)-1])
        columns = all_col[len(all_col)-1]
        shape = len(columns)
        img = np.zeros((shape, shape))
        shared_cols.append([])
        for k in range(shape):
            if columns[k] in data[i] and columns[k] in data[i+j+1]:
                img[k,k] = 1
                shared_cols[-1].append(columns[k])
        imgs.append(img)
```

```python
plt.figure(figsize = (20,20))
ax = sns.heatmap(imgs[0], annot=True, linewidths=.5, xticklabels = all_col[0], yticklabels=all_col[0])
```

## На что можно завязаться
**case:concept:name**, который является id для каждого кейса разный в каждом логе, т. е. одна и та же заявка на разных стадиях одобрения является ***разными кейсами***.

Для международных и внутренних поездок можно взять параметр **case:id** (он же **case:concept:name**) и **case:id** (он же **case:Permit id** и **case:Permit ID**) и на их основе объединить логи с логом "разрешений" по столбцам **case:dec_id_x** где x от 0 до 16. По моему предположению, это несколько заявок на одну и ту же поездку, которая состоялась. Но кол-во полей в логах с одинаковым **case:id** отличается, при этом нет никаких других уникальных полей, по которым можно было бы сопоставить данные из разных логов так, чтобы можно было размножить или сократить число одинаковых строчек

Для логов "подготовки к путешествию" и "запрос оплаты" есть поля **case:Rfp_id**, **case:RfpNumber** и **case:concept:name**, но совпадений между этими логами найдено не было

Для лога "подготовки к путешествию" и лога "разрешений" есть поля **case:Rfp_id_x**, **case:concept:name** и **case:id**, но кол-во строк с одинаковыми параметрами в перечисленных полях в логах разное
